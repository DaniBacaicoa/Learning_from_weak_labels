```python
import torch 

import pickle
import matplotlib.pyplot as plt
```

```python
Methods = ['Back','Back_conv','Back_opt','Back_opt_conv','EM','OSL']
Names = ['Backward','Backward convex','Backward optimized','Backward optimized convex',
'EM','OSL']
Folder = 'Experimental_results(0.5)'
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

Metric = ['train_loss','train_acc','test_acc']
Metric_name = ['Train loss', 'Train accuracy', 'Test accuracy']
ax_name = ['loss','acc','acc']

metric_to_use = 0
used_metric = Metric[metric_to_use]
plt.figure(figsize=(10, 6))

for e,i in enumerate(Methods):
    path = Folder + '/' + i + ".pkl"
    f = open(path,"rb")
    Results, model = pickle.load(f)
    f.close()

    Res = torch.vstack(tuple((Results[i][used_metric] for i in range(len(Results)))))
    mres = torch.mean(Res, dim=0).numpy()
    print(mres)
    sres = torch.std(Res, dim=0).numpy()
    plt.plot(mres , label = Names[e], color = colors[e])
    plt.fill_between(range(mres.shape[0]), mres-sres, mres+sres, color = colors[e], alpha=0.2)

plt.title(Metric_name[metric_to_use] + '\np=' + '0.5')
plt.xlabel('epoch')
plt.ylabel(ax_name[metric_to_use])
plt.legend()
plt.grid(True)
plt.show()

```

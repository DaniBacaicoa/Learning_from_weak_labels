```python
import matplotlib.pyplot as plt
import pickle
import torch

def plot_experiment_results(methods, names, palette, graphics, result_path, title):
    num_colors = len(methods)

    cmap = plt.get_cmap(palette)
    colors = [cmap(i / num_colors) for i in range(num_colors)]  # Normalize color indices

    plt.figure(figsize=(10, 6))
    fig, axs = plt.subplots(1, len(graphics) - 1, figsize=(12, 6))  # Adjusted to handle variable number of graphics

    for j, los in enumerate(graphics[1:]):
        for e, method in enumerate(methods):
            file = f"{result_path}/{method}.pkl"
            with open(file, "rb") as f:
                k = pickle.load(f)
                k = k['overall_results']
                series = torch.vstack(tuple((k[i][los] for i in range(len(k)))))

                mean_series = torch.mean(series, dim=0).numpy()
                std_series = torch.std(series, dim=0).numpy()

                axs[j].plot(mean_series, label=names[e], color=colors[e])
                axs[j].fill_between(range(mean_series.shape[0]), mean_series - std_series, mean_series + std_series, color=colors[e], alpha=0.4)
        
        axs[j].set_title(graphics[j+1])
        axs[j].set_xlabel('epoch')
        axs[j].set_ylabel('loss')
        if j == 1:
            axs[j].legend(bbox_to_anchor=(1.1, 1.05))
        axs[j].grid(True)
    
    fig.suptitle(title)
    plt.show()
```
